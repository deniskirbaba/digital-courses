## РџСЂРёРјРµСЂ РєР»Р°СЃС‚РµСЂРёР·Р°С†РёРё РЅР° СЂРµР°Р»СЊРЅС‹С… РґР°РЅРЅС‹С…

РЎРѕРіР»Р°СЃРЅРѕ РѕРїРёСЃР°РЅРёСЋ, РѕР±СЉРµРєС‚С‹ вЂ” Р»СЋРґРё РІРѕР·СЂР°СЃС‚Р° 18 Р»РµС‚. РљР°Р¶РґС‹Р№ РѕР±СЉРµРєС‚ РѕР±Р»Р°РґР°РµС‚ С‚СЂРµРјСЏ Р°С‚СЂРёР±СѓС‚Р°РјРё: РџРѕР» (Gender), Р РѕСЃС‚ (Height) Рё Р’РµСЃ (Weight). Р”Р»СЏ РЅР°С‡Р°Р»Р° РїРѕРґРєР»СЋС‡РёРј РЅРµРѕР±С…РѕРґРёРјС‹Рµ Р±РёР±Р»РёРѕС‚РµРєРё.

```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline

import os
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn import preprocessing
```

РРјРїРѕСЂС‚РёСЂСѓРµРј РґР°С‚Р°СЃРµС‚ РІ РґР°С‚Р°С„СЂРµР№Рј Рё РїСЂРѕРІРµСЂРёРј, РІСЃРµ Р»Рё РєРѕСЂСЂРµРєС‚РЅРѕ РёРјРїРѕСЂС‚РёСЂРѕРІР°Р»РѕСЃСЊ. 

<b>РџСЂРёРјРµС‡Р°РЅРёРµ:</b> РР·РЅР°С‡Р°Р»СЊРЅРѕ СЂРѕСЃС‚ Р±С‹Р» СѓРєР°Р·Р°РЅ РІ РґСЋР№РјР°С…, Р° РІРµСЃ -- РІ С„СѓРЅС‚Р°С…. Р”Р»СЏ СѓРґРѕР±СЃС‚РІР° РІРѕСЃРїСЂРёСЏС‚РёСЏ Р·РЅР°С‡РµРЅРёСЏ Р±С‹Р»Рё РїРµСЂРµРІРµРґРµРЅС‹ РІ СЃР°РЅС‚РёРјРµС‚СЂС‹ Рё РєРёР»РѕРіСЂР°РјРјС‹, СЃРѕРѕС‚РІРµС‚СЃС‚РІРµРЅРЅРѕ.

```python
df = pd.read_csv("height-weight_metric.csv", sep=";", decimal='.', index_col='id')
df.head()
```

РџРѕСЃРєРѕР»СЊРєСѓ РєР°Р¶РґС‹Р№ РѕР±СЉРµРєС‚ РёРјРµРµС‚ РІСЃРµРіРѕ С‚СЂРё РїСЂРёР·РЅР°РєР°, РІРїРѕР»РЅРµ РІРѕР·РјРѕР¶РЅРѕ РїРѕСЃС‚СЂРѕРёС‚СЊ РіСЂР°С„РёРє РІ 3d. Р”Р»СЏ СЌС‚РѕРіРѕ СЃРЅР°С‡Р°Р»Р° СЃРґРµР»Р°РµРј РїСЂРёР·РЅР°Рє <code>Gender</code> С‡РёСЃР»РѕРІС‹Рј (Р¶РµРЅС‰РёРЅР°Рј РїРѕСЃС‚Р°РІРёРј РІ СЃРѕРѕС‚РІРµС‚СЃС‚РІРёРµ 0, РјСѓР¶С‡РёРЅР°Рј вЂ” 1).

```python
### 3d РїРѕСЃС‚СЂРѕРµРЅРёРµ
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

sns.set(style = "darkgrid")
df['Gender'].replace('Female', 0, inplace = True)
df['Gender'].replace('Male', 1, inplace = True)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = df['Gender']
y = df['Height']
z = df['Weight']

ax.set_xlabel("Gender")
ax.set_ylabel("Height")
ax.set_zlabel("Weight")
ax.scatter(x, y, z)
plt.show()
```

РќРµС‚СЂСѓРґРЅРѕ Р·Р°РјРµС‚РёС‚СЊ, С‡С‚Рѕ РѕР±СЉРµРєС‚С‹ СЏРІРЅРѕ РјРѕР¶РЅРѕ СЂР°Р·РґРµР»РёС‚СЊ РЅР° 2 РєР»Р°СЃС‚РµСЂР° РїРѕ РїРѕР»РѕРІРѕРјСѓ РїСЂРёР·РЅР°РєСѓ. Р­С‚Рѕ Р±С‹Р»Рѕ РѕС‡РµРІРёРґРЅРѕ СЃ СЃР°РјРѕРіРѕ РЅР°С‡Р°Р»Р°, РїРѕСЌС‚РѕРјСѓ РїРѕРїСЂРѕР±СѓРµРј РЅР° РїРѕР» РїРѕРєР° С‡С‚Рѕ РІРЅРёРјР°РЅРёСЏ РЅРµ РѕР±СЂР°С‰Р°С‚СЊ (РёСЃРєР»СЋС‡РёРј РµРіРѕ РёР· РїСЂРµРґРёРєС‚РѕСЂРѕРІ).

```python
df_no_gender = df.drop(columns = 'Gender')
sns.scatterplot(data = df_no_gender, x = "Height", y = "Weight")
```

Р’С‹Р±СЂРѕСЃРѕРІ РЅРµ РІРёРґРЅРѕ. Р”Р°РЅРЅС‹Рµ РёРјРµСЋС‚ СЂР°Р·РЅС‹Рµ РµРґРёРЅРёС†С‹ РёР·РјРµСЂРµРЅРёСЏ, СЃС‚Р°РЅРґР°СЂС‚РёР·СѓРµРј РёС…. РСЃРїРѕР»СЊР·СѓРµРј Р»РёРЅРµР№РЅСѓСЋ РЅРѕСЂРјРёСЂРѕРІРєСѓ (MinMaxScaler).

```python
stand = preprocessing.MinMaxScaler()
X = stand.fit_transform(df_no_gender)
X = pd.DataFrame(X, index = df_no_gender.index, columns = df_no_gender.columns)
sns.scatterplot(data = X, x = "Height", y = "Weight")
```

РўРµРїРµСЂСЊ РЅСѓР¶РЅРѕ РѕРїСЂРµРґРµР»РёС‚СЊСЃСЏ СЃ РєРѕР»РёС‡РµСЃС‚РІРѕРј РєР»Р°СЃС‚РµСЂРѕРІ. РџРѕСЃС‚СЂРѕРёРј РіСЂР°С„РёРє РєР°РјРµРЅРёСЃС‚РѕР№ РѕСЃС‹РїРё.

```python
# РџСЂРѕРёР·РІРµРґРµРј РёРµСЂР°СЂС…РёС‡РµСЃРєСѓСЋ РєР»Р°СЃС‚РµСЂРёР·Р°С†РёСЋ. Р”Р»СЏ РѕРїСЂРµРґРµР»РµРЅРёСЏ СЂР°СЃСЃС‚РѕСЏРЅРёСЏ РјРµР¶РґСѓ РєР»Р°СЃС‚РµСЂР°РјРё Р±СѓРґРµРј РёСЃРїРѕР»СЊР·РѕРІР°С‚СЊ РјРµС‚РѕРґ Р’Р°СЂРґР°.
# РџРѕСЃС‚СЂРѕРёРј РіСЂР°С„РёРє РєР°РјРµРЅРёСЃС‚РѕР№ РѕСЃС‹РїРё РґР»СЏ РѕРїСЂРµРґРµР»РµРЅРёСЏ С‡РёСЃР»Р° РєР»Р°СЃС‚РµСЂРѕРІ.
link = linkage(X, 'ward', 'euclidean')
dist = link[:, 2][::-1]
idxs = range(1, len(dist) + 1)

```

```python
#Р Р°СЃСЃРјРѕС‚СЂРёРј СЃР»СѓС‡Р°Рё РґРѕ 50 РєР»Р°СЃС‚РµСЂРѕРІ


rock_line = pd.DataFrame()
rock_line[r'$K$ - С‡РёСЃР»Рѕ РєР»Р°СЃС‚РµСЂРѕРІ'] = idxs[:50]
rock_line[r'$\rho(X, X^\prime)$'] = dist[:50]
sns.lineplot(data = rock_line, x = r"$K$ - С‡РёСЃР»Рѕ РєР»Р°СЃС‚РµСЂРѕРІ", y = r"$\rho(X, X^\prime)$", markers = True, marker = 'o').set_title('Р“СЂР°С„РёРє РєР°РјРµРЅРёСЃС‚РѕР№ РѕСЃС‹РїРё')
```

Р Р°СЃСЃРјРѕС‚СЂРёРј Р±РѕР»РµРµ РґРµС‚Р°Р»СЊРЅРѕ СЃР»СѓС‡Р°Рё РґРѕ 10 РєР»Р°СЃС‚РµСЂРѕРІ.

```python
rock_line = pd.DataFrame()
rock_line[r'$K$ - С‡РёСЃР»Рѕ РєР»Р°СЃС‚РµСЂРѕРІ'] = idxs[:10]
rock_line[r'$\rho(X, X^\prime)$'] = dist[:10]
lnplt = sns.lineplot(data = rock_line, x = r"$K$ - С‡РёСЃР»Рѕ РєР»Р°СЃС‚РµСЂРѕРІ", y = r"$\rho(X, X^\prime)$", markers = True, marker = 'o').set_title('Р“СЂР°С„РёРє РєР°РјРµРЅРёСЃС‚РѕР№ РѕСЃС‹РїРё')
lnplt = lnplt.get_figure()    
lnplt.savefig('elbow_hie.png', dpi=400)
```

РњРѕР¶РЅРѕ Р·Р°РјРµС‚РёС‚СЊ, С‡С‚Рѕ РіСЂР°С„РёРє РЅР°С‡РёРЅР°РµС‚ Р±С‹С‚СЊ Р±РѕР»РµРµ РїРѕР»РѕРіРёРј, РЅР°С‡РёРЅР°СЏ СЃ РѕС‚РјРµС‚РєРё РІ С€РµСЃС‚СЊ РєР»Р°СЃС‚РµСЂРѕРІ. РўР°РєР¶Рµ РЅРµРїР»РѕС…РёРµ РєР°РЅРґРёРґР°С‚С‹ вЂ” С‡РµС‚С‹СЂРµ Рё РґРІР°. РџРѕСЃРєРѕР»СЊРєСѓ РґРІР° вЂ” СЃР»РёС€РєРѕРј РѕС‡РµРІРёРґРЅС‹Р№ РІР°СЂРёР°РЅС‚, СЂР°СЃСЃРјРѕС‚СЂРёРј СЃР»СѓС‡Р°Рё С€РµСЃС‚Рё Рё С‡РµС‚С‹СЂРµС… РєР»Р°СЃС‚РµСЂРѕРІ.

```python
import matplotlib
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

#РџРѕСЃС‚СЂРѕРµРЅРёРµ РґРµРЅРґСЂРѕРіСЂР°РјРјС‹ Р·Р°РЅРёРјР°РµС‚ РЅРµРєРѕС‚РѕСЂРѕРµ РІСЂРµРјСЏ. РЎС‚РѕРёС‚ РЅР°Р±СЂР°С‚СЊСЃСЏ С‚РµСЂРїРµРЅРёСЏ.

fig_1 = plt.figure(figsize=(12,12))
dn = dendrogram(link,
                color_threshold=dist[4],
                orientation = "right",
                leaf_rotation=0.)
```

```python
#Р”РѕР±Р°РІРёРј РІ РґР°РЅРЅС‹Рµ РіСЂР°С„Сѓ "РєР»Р°СЃС‚РµСЂ" Рё РїРѕСЃРјРѕС‚СЂРёРј
df_no_gender['РЎluster'] = fcluster(link, 6, criterion='maxclust')
df_no_gender['РЎluster'] = df_no_gender['РЎluster'] - 1
df_no_gender.head()
df_no_gender.groupby('РЎluster').mean()
```
